# Experiment: 2+2=4

**Claim:** The expression 2+2 evaluates to 4

**Type:** Contract

**Verdict:** SURVIVED

## Kill Criteria

1. Dies if 2+2 != 4 in Python
2. Dies if 2+2 != 4 in integer arithmetic
3. Dies if 2+2 == 5 (Orwellian test)

````python
# Test 1: Direct evaluation
result = 2 + 2
print(f"Test 1 - Direct evaluation: 2+2 = {result}")
print(f"  Criterion: Dies if 2+2 != 4")
print(f"  Result: {'KILLED' if result != 4 else 'SURVIVED'}")
````

````python
# Test 2: Comparison
is_four = (2 + 2 == 4)
print(f"Test 2 - Comparison: (2+2 == 4) = {is_four}")
print(f"  Criterion: Dies if False")
print(f"  Result: {'KILLED' if not is_four else 'SURVIVED'}")
````

````python
# Test 3: Orwellian test
is_five = (2 + 2 == 5)
print(f"Test 3 - Orwellian: (2+2 == 5) = {is_five}")
print(f"  Criterion: Dies if 2+2 == 5")
print(f"  Result: {'KILLED' if is_five else 'SURVIVED'}")
````

````python
# Type verification
print(f"type(2+2) = {type(2+2)}")
print(f"2+2 is int: {isinstance(2+2, int)}")
````
